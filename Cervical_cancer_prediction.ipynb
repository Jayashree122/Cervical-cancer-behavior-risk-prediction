```python
x=d.iloc[:,:-1]
y=d.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
a=RandomForestClassifier(n_estimators=100,criterion='gini')
a.fit(xtrain,ytrain)
```

```python
ypred=a.predict(xtest)
```

```python
from sklearn.metrics import confusion_matrix as cm,accuracy_score as ac,classification_report as cr
print("Accuracy score =",ac(ytest,ypred))
```

```python
print("Confusion matrix =",cm(ytest,ypred))
```

```python
print("Classification report =",cr(ytest,ypred))
```
